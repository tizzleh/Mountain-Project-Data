```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import seaborn as sns; sns.set(color_codes=True)

years = mdates.YearLocator() # Every year
months = mdates.MonthLocator() # Every month
years_fmt = mdates.DateFormatter('%Y')
mp_data = pd.read_csv("mp_data_clean.csv")
pd.to_datetime(mp_data.Date)
mp_data['Date'] = pd.to_datetime(mp_data['Date'])
mstd = mp_data.rolling(20).mean()



print(mp_data['Date'])
print(mstd)
mp_data.plot(x="Date", y="New Users", kind='line', lw=.25)
re = mstd.plot(lw=.5)


re.annotate('Oscillations',
            xy=(4200, 60), xycoords='data',
            xytext=(0.5, 0.55), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='top')

re.annotate('',
            xy=(5000, 180), xycoords='data',
            xytext=(0.5, 0.55), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='top')


plt.savefig("mp_total.svg")
```
